Histograms

```python
def func_plot(x_axis,color_p1,color_p2):
    fig, (ax1, ax2) = plt.subplots(nrows=1, ncols=2, figsize=(20, 5), facecolor='#333333')

    df1=df[x_axis].value_counts().sort_values(ascending=False)
    colors1 = cm.get_cmap(color_p1, len(df1))
    df1.plot(kind='bar', ax=ax1, color=colors1(range(len(df1))) )
    ax1.set_title(x_axis + ' Histogram')
    ax1.title.set_color('white')
    ax1.set_xlabel(x_axis,color='white')

    # Loop through the texts of axes and set their color to white
    for axis_texts in ax1.get_xticklabels() + ax1.get_yticklabels():
        axis_texts.set_color('white')
    
    df2 = pd.DataFrame(df.groupby([x_axis])['price'].mean().sort_values(ascending=False))
    df2.plot(kind='bar', ax=ax2, color=color_p2)
    ax2.set_title(x_axis + ' vs Mean Price')
    ax2.title.set_color('white')
    ax2.set_xlabel(x_axis,color='white')

    # Loop through the texts of axes and set their color to white
    for axis_texts in ax2.get_xticklabels() + ax2.get_yticklabels():
        axis_texts.set_color('white')

    plt.show()
```

PIE CHARTS

```python
def plot_pie_chart(column_name, color_palette, explode_index=None):
    data = df[column_name].value_counts()

    # Extract the labels and sizes from the data
    labels = data.index.tolist()
    sizes = data.values.tolist()

    # Generate a color map with the same length as the number of categories
    colors = plt.cm.get_cmap(color_palette, len(labels))

    # Create the pie chart with exploding a slice
    fig, ax = plt.subplots(figsize=(16, 8))
    explode = [0] * len(labels)
    if explode_index is not None:
        explode[explode_index] = 0.2  # Explode the specified slice
    wedges, texts = ax.pie(sizes, colors=colors(range(len(labels))), radius=1.2, startangle=90, explode=explode)

    # Set border color for each wedge
    for wedge in wedges:
        wedge.set_edgecolor('brown')
    
    # Add percentage values to the legend
    percentages = [f'{size/sum(sizes)*100:.1f}%' for size in sizes]
    legend_labels = [f'{label} - {percentage}' for label, percentage in zip(labels, percentages)]
    ax.legend(wedges, legend_labels, loc='upper left', fontsize=8)

    # Set title and legend
    ax.set_title('Distribution of ' + column_name)

    # Set aspect ratio to make the pie chart a circle
    ax.axis('equal')

    # Show the chart
    plt.show()
```

```python
plot_pie_chart('CompanyName','magma',0)
```

```python
plot_pie_chart('enginetype','jet',0)
```

```python
plot_pie_chart('carbody','rainbow',0)
```

```python
plot_pie_chart('fuelsystem','viridis_r',0)
```

SCATTER PLOTS

The Scatter plot here mainly tells us about the corelation between the two specific columns

```python
def plot_scatter(column1, column2, color_palette):
    # Extract the data from the columns
    data1 = df[column1]
    data2 = df[column2]
    
    # Find the correlation
    correlation = data1.corr(data2)
    print(correlation)
    
    if correlation > 0:
        relation = "Positive"
    elif correlation < 0:
        relation = "Negative"
    else:
        relation = "No Relation"

    # Set the color palette
    colors = sns.color_palette(color_palette, n_colors=len(data1))

    # Create a scatter plot
    plt.figure(figsize=(10, 6))
    plt.scatter(data1, data2, c=colors)

    # Set labels and title
    plt.xlabel(column1)
    plt.ylabel(column2)
    plt.title(f'{relation} Correlation between {column1} and {column2}')

    # Show the plot
    plt.show()
```
